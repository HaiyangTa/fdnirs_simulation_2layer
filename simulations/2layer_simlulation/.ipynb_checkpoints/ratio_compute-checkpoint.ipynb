```python
import numpy as np
```

```python

def cw_depth_sensitivity(d, rho,
                         mu_a, mu_sp, n=1.4,
                         include_image=True):
    D = 1/(3*(mu_a + mu_sp))
    mu_eff = np.sqrt(mu_a/D)
    c0 = 3e11
    Re = 0.493
    A  = (1+Re)/(1-Re)
    z_b = 2*A*D
    
    r1 = np.sqrt(rho**2 + d**2)
    term = np.exp(-mu_eff*r1)/r1
    
    if include_image:
        r2 = np.sqrt(rho**2 + (2*z_b + d)**2)
        term -= np.exp(-mu_eff*r2)/r2
    
    return term/(4*np.pi*D)

def normalized_S(d, rho, mu_a, mu_sp):
    D_max = 100  # mm
    dd = np.linspace(0, D_max, 2001)
    Sd = cw_depth_sensitivity(dd, rho, mu_a, mu_sp)
    norm = np.trapz(Sd, dd)
    return cw_depth_sensitivity(d, rho, mu_a, mu_sp) / norm
    
rho = 45  # mm
mu_a, mu_sp = 0.01, 1.0  # mm^-1
d = np.linspace(0, 100, 100) # depth of the tissue in mm. 
S_norm = normalized_S(d, rho, mu_a, mu_sp)
```
