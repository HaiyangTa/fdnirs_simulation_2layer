```python
"""
Combine all files from N source folders into one destination folder.

• Adjust ROOT_DIR  → the directory that contains the 10 folders
• Adjust DST_DIR   → the new folder that will gather all 1 000 files
• If you want to *copy* instead of *move* just switch
      shutil.move  →  shutil.copy2
"""

import shutil, os, pathlib

# ------------------------------------------------------------------
ROOT_DIR = pathlib.Path(r"C:/Users/Haiya/Downloads/testset/")   # contains the 10 folders
DST_DIR  = ROOT_DIR / "combined_files"        # new folder to create
# ------------------------------------------------------------------

DST_DIR.mkdir(exist_ok=True)                  # make destination once

moved = 0
for folder in ROOT_DIR.iterdir():             # loop over the 10 folders
    if folder.is_dir():
        for file in folder.iterdir():         # loop over the 100 files
            if file.is_file():
                target = DST_DIR / file.name
                # ensure unique names; add a prefix if duplicate
                if target.exists():
                    target = DST_DIR / f"{folder.name}_{file.name}"
                shutil.move(str(file), str(target))   # ← or .copy2(...)
                moved += 1

print(f"Done – {moved} files moved to {DST_DIR}")
```
